```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def euclidean_distance(p, q):
    # using numpy's linalg.norm() function to calculate Euclidean distance
    return np.linalg.norm(p - q)
```

```python
def plot_points(p, q, distance):
    plt.figure(figsize=(15,9))

    if p.size == 2 and q.size == 2: # for 2D points
        ax1 = plt.subplot(1, 2, 1)
        ax1.scatter([p[0], q[0]], [p[1], q[1]], c='r')
        ax1.plot([p[0], q[0]], [p[1], q[1]], 'k-') # line between points
        ax1.plot([p[0], q[0]], [p[1], p[1]], 'b--') # x-axis component
        ax1.plot([q[0], q[0]], [p[1], q[1]], 'g--') # y-axis component
        ax1.set_xlabel('X')
        ax1.set_ylabel('Y')
        ax1.annotate(f'P({p[0]}, {p[1]})', (p[0], p[1]), textcoords="offset points", xytext=(-10,-10), ha='center', fontsize=12, color='blue')
        ax1.annotate(f'Q({q[0]}, {q[1]})', (q[0], q[1]), textcoords="offset points", xytext=(10,10), ha='center', fontsize=12, color='blue')
        ax1.set_title(f'Euclidean distance between points: {distance:.2f}', fontsize=15)
        ax1.grid(True)

    elif p.size == 3 and q.size == 3: # for 3D points
        ax2 = plt.subplot(1, 2, 2, projection='3d')
        ax2.scatter([p[0], q[0]], [p[1], q[1]], [p[2], q[2]], c='b')
        ax2.plot([p[0], q[0]], [p[1], q[1]], [p[2], q[2]], 'k-') # line between points
        ax2.plot([p[0], q[0]], [p[1], p[1]], [p[2], p[2]], 'r--') # x-axis component
        ax2.plot([q[0], q[0]], [p[1], q[1]], [p[2], p[2]], 'g--') # y-axis component
        ax2.plot([q[0], q[0]], [q[1], q[1]], [p[2], q[2]], 'b--') # z-axis component
        ax2.set_xlabel('X')
        ax2.set_ylabel('Y')
        ax2.set_zlabel('Z')
        ax2.text(p[0], p[1], p[2], f'P({p[0]}, {p[1]}, {p[2]})', fontsize=12, color='blue')
        ax2.text(q[0], q[1], q[2], f'Q({q[0]}, {q[1]}, {q[2]})', fontsize=12, color='blue')
        ax2.set_title(f'Euclidean distance between points: {distance:.2f}', fontsize=15)

    #plt.suptitle(f'Euclidean distance between points: {distance:.2f}', fontsize=16)
    plt.show()
```

```python
# Testing the functions with 2D points
p = np.array([2, 3])
q = np.array([4, 5])
distance = euclidean_distance(p, q)
plot_points(p, q, distance)
```

```python
# Testing the functions with 3D points
p = np.array([2, 3, 1])
q = np.array([4, 5, 6])
distance = euclidean_distance(p, q)
plot_points(p, q, distance)
```
